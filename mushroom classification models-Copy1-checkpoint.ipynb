```python
import numpy as np 
import pandas as pd
import warnings
warnings.simplefilter("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score, accuracy_score
from sklearn.tree import export_graphviz
from sklearn.metrics import precision_recall_curve, auc
from sklearn.model_selection import cross_val_predict, cross_val_score
```

```python
df = pd.read_csv("mushrooms.csv")
```

```python
df.head()
```

```python
count = df['class'].value_counts()
sns.set_context("notebook", font_scale=1)
plt.figure(figsize=(8,7))
sns.barplot(x=count.index, y=count.values, alpha=0.8, palette="bwr")
plt.ylabel('Count')
plt.xlabel('Class')
plt.title('Number of p and e mushrooms')
plt.show()
```

```python
df['class'].value_counts()
```

```python
df.shape
```

```python
X = df.drop(['class'], axis=1)  
y = df["class"]

X.head()
```

```python
labelencoder_x=LabelEncoder()
for column in X.columns:
    X[column] = labelencoder_x.fit_transform(X[column])
    
labelencoder_y=LabelEncoder()
y = labelencoder_y.fit_transform(y)
```

```python
X.head()
```

```python
X=pd.get_dummies(X,columns=X.columns,drop_first=True)
X.head()
```

```python
plt.figure(figsize=(16,12))
sns.heatmap(X.corr(),linewidths=.0,cmap="Blues")
plt.yticks(rotation=0);
#plt.savefig("corr2.jpeg", format='jpeg', dpi=300, bbox_inches='tight')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X2 = sc.fit_transform(X)
X_test = sc.transform(X_test)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.3, random_state=42)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_train_pca=pca.fit_transform(X_train) 
X_test_pca=pca.transform(X_test)
print("Original shape:",X_train.shape)
print('Shape of PCA data:',X_train_pca.shape)
```

```python
#LOGISTIC REGRESSION
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
```

```python
# Scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)
X_test_scaled = scaler.transform(X_test_pca)  
```

```python
lr = LogisticRegression(random_state=42, C=1.5, solver='liblinear', penalty='l2')  # C adjusted for accuracy
lr.fit(X_train_scaled, y_train)
```

```python
# Training Results
print("TRAINING RESULTS:\n")
train_predictions = lr.predict(X_train_scaled)
train_accuracy = accuracy_score(y_train, train_predictions)
print('Accuracy Score (Train): {0:.4f}\n'.format(train_accuracy))
print('Classification Report (Train):\n{}\n'.format(classification_report(y_train, train_predictions)))
print('Confusion Matrix (Train):\n{}\n'.format(confusion_matrix(y_train, train_predictions)))
```

```python
# Cross-validation on Training Set
res = cross_val_score(lr, X_train_scaled, y_train, cv=10, n_jobs=-1, scoring='accuracy')
print('Average Accuracy (CV on Train Set):\t{0:.4f}\n'.format(res.mean()))
print('Standard Deviation (CV on Train Set):\t{0:.4f}'.format(res.std()))
```

```python
# Test Results
print("TEST RESULTS:\n")
test_predictions = lr.predict(X_test_scaled)
test_accuracy = accuracy_score(y_test, test_predictions)
print('Accuracy Score (Test): {0:.4f}\n'.format(test_accuracy))
print('Classification Report (Test):\n{}\n'.format(classification_report(y_test, test_predictions)))
print('Confusion Matrix (Test):\n{}\n'.format(confusion_matrix(y_test, test_predictions)))
```

```python
#Visualizing the Decision Boundary for Training Set
sns.set_context(context='notebook', font_scale=1.5)
plt.figure(figsize=(16, 9))
X_set, y_set = X_train_scaled, y_train
step_size = 0.05  # Adjust this step size for faster plotting
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=step_size),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=step_size))

Z = lr.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape)

plt.contourf(X1, X2, Z, alpha=0.3, cmap=ListedColormap(('lightcoral', 'mediumslateblue')))

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('lightcoral', 'mediumslateblue'))(i), label=j)

plt.title("Logistic Regression Training Set")
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.legend()
plt.savefig("lr_pci_tr_75_percent.jpeg", format='jpeg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import ListedColormap

# Scaling the data first
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_pca)  
X_test_scaled = scaler.transform(X_test_pca)         

# Define a Decision Tree as the base estimator for Bagging
dtree = DecisionTreeClassifier(random_state=42, max_depth=5)  

# Use Bagging with Decision Trees
bagging_clf = BaggingClassifier(estimator=dtree,  # Pass the estimator here
                               n_estimators=20,  # Reduced number of trees (default is 10)
                               max_features=0.5,  # Limit the number of features per tree
                               random_state=42,   # Random seed for reproducibility
                               n_jobs=-1)         # Use all CPU cores
```

```python
# Train the model
bagging_clf.fit(X_train_scaled, y_train)

# Check the accuracy after training
train_accuracy = accuracy_score(y_train, bagging_clf.predict(X_train_scaled))
test_accuracy = accuracy_score(y_test, bagging_clf.predict(X_test_scaled))

print(f"Training Accuracy: {train_accuracy:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

print("TRAINING RESULTS:\n")
print('Accuracy Score: {0:.4f}\n'.format(train_accuracy))
print('Classification Report:\n{}\n'.format(classification_report(y_train, bagging_clf.predict(X_train_scaled))))
print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_train, bagging_clf.predict(X_train_scaled))))
```

```python
# Cross-validation score (to ensure consistency)
res = cross_val_score(bagging_clf, X_train_scaled, y_train, cv=10, n_jobs=-1, scoring='accuracy')
print('Average Accuracy:\t{0:.4f}\n'.format(res.mean()))
print('Standard Deviation:\t{0:.4f}'.format(res.std()))
```

```python
# Test results
print("TEST RESULTS:\n")
print('Accuracy Score: {0:.4f}\n'.format(test_accuracy))
print('Classification Report:\n{}\n'.format(classification_report(y_test, bagging_clf.predict(X_test_scaled))))
print('Confusion Matrix:\n{}\n'.format(confusion_matrix(y_test, bagging_clf.predict(X_test_scaled))))
```

```python
# Visualizing the decision boundary for Bagging Decision Tree
sns.set_context(context='notebook', font_scale=1.5)
plt.figure(figsize=(16, 9))

X_set, y_set = X_train_scaled, y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

plt.contourf(X1, X2, bagging_clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.6, cmap=ListedColormap(('lightcoral', 'mediumslateblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('lightcoral', 'mediumslateblue'))(i), label=j)

plt.title("Bagging Decision Tree Training Set")
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.savefig("bagging_dt_pci_tr_80_percent_accuracy.jpeg", format='jpeg', dpi=300, bbox_inches='tight')
plt.legend()
```

```python
# Visualizing the decision boundary on the test set using Bagging Decision Tree
sns.set_context(context='notebook', font_scale=1.5)
plt.figure(figsize=(16, 9))

X_set, y_set = X_test_scaled, y_test  # Use the scaled test set for visualization
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1, stop=X_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1, stop=X_set[:, 1].max() + 1, step=0.01))

# Predict the classes for the mesh grid to plot decision boundaries
plt.contourf(X1, X2, bagging_clf.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.6, cmap=ListedColormap(('lightcoral', 'mediumslateblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

# Plot the actual points from the test set
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c=ListedColormap(('lightcoral', 'mediumslateblue'))(i), label=j)

plt.title("Bagging Decision Tree Test Set")
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')

# Save the figure
plt.savefig("bagging_dt_pci_ts_accuracy_80_percent.jpeg", format='jpeg', dpi=300, bbox_inches='tight')

# Add legend
plt.legend()
plt.show()
```
